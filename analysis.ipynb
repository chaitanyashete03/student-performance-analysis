```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_data.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

plt.figure()
plt.hist(df["Marks"], bins=6)

plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()
```

# insight 
Most students fall into the mid-range of marks, indicating a generally balanced academic performance.
The absence of extreme peaks suggests there is no heavy concentration of failures or toppers, pointing toward consistent overall outcomes.

```python
import matplotlib.pyplot as plt

avg_marks = df.groupby("Gender")["Marks"].mean()

plt.figure()
bars = plt.bar(avg_marks.index, avg_marks.values)

plt.title("Average Marks by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Marks")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{height:.1f}",
        ha="center",
        va="bottom"
    )

plt.show()
```

# insight
The chart shows that male students have a higher average marks compared to female students.
The exact values displayed on the chart confirm this difference, although the gap is moderate, indicating comparable performance levels across genders.

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

plt.figure(figsize=(7,5))

plt.scatter(
    df["Study_Hours"],
    df["Marks"],
    s=80,         
    alpha=0.8       
)

plt.title("Relationship Between Study Hours and Marks", fontsize=14)
plt.xlabel("Study Hours", fontsize=11)
plt.ylabel("Marks", fontsize=11)

plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))

plt.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

# insight
The scatter plot shows a clear positive relationship between study hours and marks.
As study hours increase, student performance consistently improves, indicating that study time is a strong contributor to academic outcomes.
The pattern suggests correlation rather than randomness, with no significant outliers disrupting the trend.
